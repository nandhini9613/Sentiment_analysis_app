## Importing Dataset

```python
import pandas as pd

# Load the CSV file
data = pd.read_csv('IMDB Dataset.csv')  # <- Change 'your_file.csv' to your actual CSV file path

# Show first few rows
print(data.head())
```

CLeaning

```python
import re

# Define a function to clean the text
def clean_text(text):
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = text.lower()                # Convert to lowercase
    return text

# Apply the cleaning
data['cleaned_review'] = data['review'].apply(clean_text)

# Check cleaned reviews
print(data[['review', 'cleaned_review']].head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer

# Create input (X) and output (y)
X = data['cleaned_review']
y = data['sentiment']

# Convert text to numbers
vectorizer = TfidfVectorizer()
X_vectorized = vectorizer.fit_transform(X)

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)
```

### Accuracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predict on test set
y_pred = model.predict(X_test)

# Evaluate the model
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Testing

```python
# New sentence
new_text = ["Very good movie"]

# Clean the new text
new_text_clean = [clean_text(text) for text in new_text]

# Vectorize it
new_text_vectorized = vectorizer.transform(new_text_clean)

# Predict
prediction = model.predict(new_text_vectorized)

print("Sentiment:", prediction[0])
```

### Model Creation

```python
import joblib

# Save the trained model
joblib.dump(model, 'sentiment_model.pkl')

# Save the vectorizer
joblib.dump(vectorizer, 'vectorizer.pkl')

print("Model and Vectorizer saved successfully!")
```
